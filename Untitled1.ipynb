```python
# Visualize distribution of numerical columns to detect outliers
import matplotlib.pyplot as plt
Numeric_Column ='Width'
# Example for boxplot of a column
plt.boxplot(data[Numeric_Column])
plt.show()

# Determine threshold for outliers and handle them appropriately
# Example: Remove outliers using z-score
from scipy import stats

z_scores = stats.zscore(data[Numeric_Column])
threshold = 3
outliers = (z_scores > threshold) | (z_scores < -threshold)
data = data[~outliers]
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data creation
data = {
    'LB': np.random.normal(loc=120, scale=10, size=100),  # Sample numerical variable
    'AC': np.random.randint(0, 2, size=100),  # Sample categorical variable
    'FM': np.random.normal(loc=50, scale=5, size=100),  # Sample numerical variable
    'UC': np.random.randint(0, 2, size=100),  # Sample categorical variable
    'DL': np.random.normal(loc=150, scale=20, size=100),  # Sample numerical variable
    'DS': np.random.randint(0, 2, size=100),  # Sample categorical variable
    'DP': np.random.normal(loc=100, scale=15, size=100),  # Sample numerical variable
    'ASTV': np.random.normal(loc=50, scale=10, size=100),  # Sample numerical variable
    'MSTV': np.random.normal(loc=1, scale=0.1, size=100),  # Sample numerical variable
    'ALTV': np.random.normal(loc=50, scale=10, size=100),  # Sample numerical variable
    'MLTV': np.random.normal(loc=1, scale=0.1, size=100),  # Sample numerical variable
    'Width': np.random.normal(loc=100, scale=10, size=100),  # Sample numerical variable
    'Tendency': np.random.randint(-1, 2, size=100),  # Sample categorical variable
    'NSP': np.random.randint(1, 4, size=100)  # Sample categorical variable
}

# Creating DataFrame
df = pd.DataFrame(data)

# Task 2: Statistical Summary
summary = df.describe()
print("Statistical Summary:")
print(summary)
print()

# Task 3: Data Visualization

# Histogram for a single numerical variable
numerical_variable_name = 'LB'
sns.histplot(df[numerical_variable_name], bins=20, kde=True)
plt.xlabel('Numerical Variable')
plt.ylabel('Frequency')
plt.title('Histogram of ' + numerical_variable_name)
plt.show()

# Boxplot for multiple numerical variables
sns.boxplot(data=df[['LB', 'FM']])
plt.xlabel('Variable')
plt.ylabel('Value')
plt.title('Boxplot of Numerical Variables')
plt.show()

# Bar chart for a categorical variable
sns.countplot(data=df, x='AC')
plt.xlabel('Categories')
plt.ylabel('Frequency')
plt.title('Bar Chart of Categorical Variable')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()

# Scatter plot for two numerical variables
sns.scatterplot(data=df, x='LB', y='FM')
plt.xlabel('Numerical Variable 1')
plt.ylabel('Numerical Variable 2')
plt.title('Scatter Plot of Numerical Variables')
plt.show()

# Correlation heatmap
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Pair plot for multiple pairwise comparisons
sns.pairplot(df[['LB', 'FM', 'DL', 'DP', 'ASTV']])
plt.title('Pair Plot')
plt.show()

# Violin plot for numerical variable across levels of a categorical variable
sns.violinplot(data=df, x='Tendency', y='LB')
plt.xlabel('Categories')
plt.ylabel('Numerical Variable')
plt.title('Violin Plot of ' + numerical_variable_name + ' across Categories')
plt.xticks(rotation=45)
plt.show()
```
